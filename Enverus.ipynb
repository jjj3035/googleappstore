```python
import pandas as pd

file = "googleplaystore.csv"
```

```python
file_df = pd.read_csv(file)
```

```python
file_df.head()
```

```python
file_df.info()
```

```python
file_df.describe()
```

```python
file_df["App"].value_counts()
```

```python
file_df.dropna(axis=0, how='any',subset=["Rating"], inplace=True)
```

```python
file_df.drop_duplicates(subset="App",inplace=True)
```

```python
file_df.info()
```

```python
file_df.describe()
```

```python
file_df["App"].value_counts()
```

```python
file_df["Price"].unique()
```

```python
file_df.drop(file_df.index[file_df['Price'] == "Everyone"], inplace=True)
```

```python
file_df[file_df.columns[7:8]] = file_df[file_df.columns[7:8]].replace('[\$,]', '', regex=True).astype(float)
```

```python
file_df["Installs"].unique()
```

```python
file_df[file_df.columns[5:6]] = file_df[file_df.columns[5:6]].replace('[\+,]', '', regex=True).astype(float)
```

```python
file_df["Gross Revenue"] = file_df["Price"] * file_df["Installs"]
```

```python
file_df.head()
```

```python
file_df = file_df.sort_values("Gross Revenue", ascending=False)
file_df.head(3)
```
